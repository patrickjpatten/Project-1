```python
# Important Note, we are running the code second times here for the presentation, if you want to run the code 
# with out a hitch,You should get Api key from https://developer.sportradar.com/ then create application and choose Nba
# They will give it to you Api key free but you can use only 1000 times every month that's the why we called them 
# and they Upgraded our Api key limit 3000.
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import scipy.stats as st
from scipy.stats import linregress
from config import apikey
from pprint import pprint
```

```python
# URL for GET requests to retrieve NBA season 2018 game ids from the season url
url = "https://api.sportradar.us/nba/trial/v7/en/games/2018/reg/schedule.json?api_key="+ apikey
# Print the response object to the console
response = requests.get(url).json()
print(json.dumps(response, indent=4, sort_keys=True))
```

```python
# Create a response for Game Id's, Create a empty list & For Loop, Filling up for gameids list 2018
gameids = response["games"]

gameids_list2018 = []

for gameid in gameids:
    gameids_list2018.append(gameid["id"])
gameids_list2018
```

```python
# Create a DataFrame for season 2018
NBA2018_df = pd.DataFrame({"GameID": gameids_list2018,})
NBA2018_df
```

```python
# We used to same way for collect to Nba Season 2019 Data
url2 = "https://api.sportradar.us/nba/trial/v7/en/games/2019/reg/schedule.json?api_key="+ apikey

response = requests.get(url2).json()
gameids  = response["games"]

gameids_list2019 = []

for gameid in gameids:
    gameids_list2019.append(gameid["id"])

NBA2019_df = pd.DataFrame({"GameID": gameids_list2019,})
NBA2019_df
```

```python
NBAALL_df = NBA2018_df.append(NBA2019_df)

NBAALL_df
```

```python
# Export file as a CSV 
NBAALL_df.to_csv("../Output/NBAALL_df.csv", index=False, header=True)
```

```python
# Now we have the all games id
# Starting the get game info from game id
NBA_Game_Info = "../Output/NBAALL_df.csv"
# Not every CSV requires an encoding, but be aware this can come up
NBA_Game_InfoDF = pd.read_csv(NBA_Game_Info, encoding="ISO-8859-1")
# Show just the header
NBA_Game_InfoDF 
```

```python
# set up lists to hold response data
gameids_list = NBA_Game_InfoDF["GameID"]
gameids_list
```

```python
# Nba Games Info search base url
boxscoreurl = "http://api.sportradar.us/nba/trial/v7/en/games/"
endboxscoreurl = "/summary.json?api_key=" + apikey

# set up lists to hold response data
gameids_list = NBA_Game_InfoDF["GameID"]
# game Id
gameid=[]
# Date
date = []
# City
city = []
# Country
country = []
# hometeam
hometeam = []
# Away Team
awayteam = []
#Attendance
attendances = []
#home then points
homepointstotals = []
#scoring, 0, points
home1stquarterpoints = []
#home then statistics
home_field_goals_pcts = []
home_three_points_pcts = []
home_free_throws_pcts = []
#away then points
awaypointstotals = []
#scoring, 0, points
away1stquarterpoints = []
#away then statistics
away_field_goals_pcts = []
away_three_points_pcts = []
away_free_throws_pcts = []
```

```python
# loop through game ids, make requests and parse and append
for gameids in gameids_list:
    target_url = boxscoreurl + gameids + endboxscoreurl
    boxscore_response = requests.get(target_url)
    time.sleep(1.1)
    boxscore_response_json = boxscore_response.json()

 
    try:
        print(f"{gameids} found! Appending stats"),
        gameid.append(boxscore_response_json['id']),
        date.append(boxscore_response_json["scheduled"]),
        city.append(boxscore_response_json["venue"]["city"]),
        country.append(boxscore_response_json["venue"]["country"]),
        hometeam.append(boxscore_response_json["home"]["name"]),
        awayteam.append(boxscore_response_json["away"]["name"]),
        attendances.append(boxscore_response_json['attendance']),
        homepointstotals.append(boxscore_response_json['home']['points']),
        home1stquarterpoints.append(boxscore_response_json['home']['scoring'][0]['points']),
        home_field_goals_pcts.append(boxscore_response_json['home']['statistics']['field_goals_pct']),
        home_three_points_pcts.append(boxscore_response_json['home']['statistics']['three_points_pct']),
        home_free_throws_pcts.append(boxscore_response_json['home']['statistics']['free_throws_pct']),
        awaypointstotals.append(boxscore_response_json['away']['points']),
        away1stquarterpoints.append(boxscore_response_json['away']['scoring'][0]['points']),
        away_field_goals_pcts.append(boxscore_response_json['away']['statistics']['field_goals_pct']),
        away_three_points_pcts.append(boxscore_response_json['away']['statistics']['three_points_pct']),
        away_free_throws_pcts.append(boxscore_response_json['away']['statistics']['free_throws_pct'])
        
    # Handle exceptions for a game info that is not available in the NBA API
    except:
        # Append null values
        print("Character not found")
        pass  
```

```python
NBA2_df = pd.DataFrame({
    'Attendance':attendances,
    "Home Score": homepointstotals,
    "Home 1st Q":home1stquarterpoints,
    "Home FG%":home_field_goals_pcts,
    "Home 3pt%":home_three_points_pcts,
    "Home FT%":home_free_throws_pcts,
    "Away Score":awaypointstotals,
    "Away 1st Q":away1stquarterpoints,
    "Away FG%":away_field_goals_pcts,
    "Away 3pt%":away_three_points_pcts,
    "Away FT%":away_free_throws_pcts
})
NBA2_df.to_csv("../Output/NBA_2_df.csv", index=False, header=True)
```
